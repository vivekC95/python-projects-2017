# Creating a list of periods

```python
period_tags = sdc.select(".tombstone-container .period-name")
```

```python
period_tags
```

```python
period_names = [pd.get_text() for pd in period_tags]
```

```python
period_names = period_names
```

```python
short_descrip = sdc.select(".tombstone-container .short-desc")
```

```python
short_descs = [sd.get_text() for sd in short_descrip]
```

```python
short_descs
```

```python
temperature = sdc.select(".tombstone-container .temp")
```

```python
temps = [tm.get_text() for tm in temperature]
```

```python
temps
```

```python
Titles = sdc.select(".tombstone-container img")
titles = [tt["title"] for tt in sdc.select(".tombstone-container img")]

titles
```

# Creating a Dataframe


```python
import pandas as pd
```

```python
diction = {'Period Names': period_names,
           'Short Description':short_descs,
           'Descriptions':titles,
           'Temperature':temps}
```

```python
data = pd.DataFrame(diction)
```

```python
data
```

```python
data[(data["Period Names"].str.contains("Night")) & (data["Temperature"].str.contains("Low"))]
```

```python
data
```

```python
data['Temp'] = data['Temperature'].str.extract('(\d\d)', expand = True).astype(int)
```
